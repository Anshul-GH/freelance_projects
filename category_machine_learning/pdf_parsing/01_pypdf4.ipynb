```python
from PyPDF4 import PdfFileWriter, PdfFileReader
```

```python
pdf_path = r'\src\foo.pdf'
```

```python
root = !cd
pdf_path = str(root[0]) + pdf_path
# print(root[0])
# print(pdf_path)
```

```python
pdf_doc = PdfFileReader(open(pdf_path, 'rb'))
```

```python
# extracting the document information
print(pdf_doc.documentInfo)
```

```python
# excryption check
print(pdf_doc.isEncrypted)
```

```python
# number of pages
print(pdf_doc.numPages)
```

```python
# reading the document pagewise
pdf_page_1 = pdf_doc.getPage(0)
```

```python
pdf_page_1
```

```python
# extracting text from page
text = pdf_page_1.extractText()
```
